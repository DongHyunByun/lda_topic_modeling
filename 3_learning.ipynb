# 데이터 불러오기

```python
import pickle

with open("corpus.pickle", 'rb') as fr:
    corpus = pickle.load(fr)
    
with open("dictionary.pickle", 'rb') as fr:
    dictionary = pickle.load(fr)
```

# gensim으로 lda 모델링

```python
import gensim

# passes : 알고리즘 동작횟수
# num_topics : 토픽의수
NUM_TOPICS = 20
myPasses=15

ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=myPasses)
```

```python
# 모델 피클파일로 저장
with open('lda_model20.pickle','wb') as fw:
    pickle.dump(ldamodel, fw)
```
