# 주제 분석할 파일 업로드 및 corpus 분해

```python
myTxt=[]
with open("test.txt","r") as f:
    while True:
        line = f.readline()
        if not line: 
            break
        myTxt.append(line)

myTxt=[myTxt[0].split()]
```

```python
print(myTxt)
```

```python
dictionary = corpora.Dictionary(myTxt)
corpus = [dictionary.doc2bow(text) for text in myTxt]
```

```python
print(dictionary)
print(corpus)
```

# 주제확인

```python
topics=model_20topics[corpus[0]]
```

```python
# 상위 5개 추출
def second(L):
    return L[1]

cate={0: "스포츠,운동", 1: "사회,문화,국가", 2: "인간,인문학,소설", 3: "해외,유럽", 4: "일본", 5: "한국지리", 6: "연구", 7: "기업,사업",
      8: "전쟁,군사", 9: "서양,문화,예술", 10: "동양지리,유적,역사", 11: "디지털,컴퓨터,인터넷", 12: "문화,방송,영화", 13: "교육", 14: "정치,법",
      15: "지역,위치", 16: "종교,중세", 17: "수학,자연", 18: "문화재,건축물", 19: "조선, 역사"}

# 상위 n개 추출 (최대15)
n=15
topics.sort(key=second, reverse=True)
print(topics)
highNTopic=[]
for i in range(n):
    highNTopic.append([cate[topics[i][0]],topics[i][1]])
```
